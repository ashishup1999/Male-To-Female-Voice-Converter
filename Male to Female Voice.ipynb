# Converting Male Voice To Female Voice

### Importing Packages

```python
import speech_recognition as sr
import pyttsx3 
import IPython.display as ipd
```

### Defining Key Functions

```python
def speachToText(audio_file):
    # Loop infinitely for user to speak infinitly.
    r = sr.Recognizer()
    while(1):    
        try:
            with sr.AudioFile(audio_file) as source:
                r.adjust_for_ambient_noise(source, duration=0.2) 
                audio = r.record(source)
                text = r.recognize_google(audio)
                text = text.lower()
                return text
        except sr.RequestError as e:
            print("Could not request results; {0}".format(e))

        except sr.UnknownValueError:
            print("unknown error occured")
```

```python
def maleVoice(text):
    engine = pyttsx3.init()
    sound=engine.getProperty('voices')
    engine.setProperty('voice',sound[0].id)
    nfile='test_male.wav'
    engine.save_to_file(text,nfile)
    engine.runAndWait()
    return nfile
```

```python
def femaleVoice(text):
    engine = pyttsx3.init()
    sound=engine.getProperty('voices')
    engine.setProperty('voice',sound[1].id)
    nfile='test_female.wav'
    engine.save_to_file(text,nfile)
    engine.runAndWait()
    return nfile
```

```python
def displayAudio(audio_file):
    return ipd.Audio(audio_file)
```

### Converter Function

```python
def convertMtoF(audio_file):
    text=speachToText(audio_file)
    return femaleVoice(text)
```

```python
#input
audio_file=input("Enter the audio file path :\n")
```

### Displaying Input Audio

```python
displayAudio(audio_file)
```

### Calling Converter Function

```python
displayAudio(convertMtoF(audio_file))
```
